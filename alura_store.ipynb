### Visualización

# 1. Generando gráfico

```python
ingreso_tienda = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

resumen1 = pd.DataFrame({
                        'N° de Tienda' : [1, 2, 3, 4],
                        'Ingresos por tienda' : [ingreso_tienda, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
})

resumen1
```

```python
resumen1.plot(kind = 'bar', x = 'N° de Tienda', y = 'Ingresos por tienda', figsize = (12,5), color = 'green')
```

```python
df_tienda = tienda['Categoría del Producto'].value_counts().sort_values()
df_tienda.plot(kind = 'barh', figsize = (12,5), color = 'blue')
```

```python
df_tienda2 = tienda2['Categoría del Producto'].value_counts().sort_values()
df_tienda2.plot(kind = 'barh', figsize = (12,5), color = 'blue')
```

```python
df_tienda3 = tienda3['Categoría del Producto'].value_counts().sort_values()
df_tienda3.plot(kind = 'barh', figsize = (12,5), color = 'blue')
```

```python
df_tienda4 = tienda4['Categoría del Producto'].value_counts().sort_values()
df_tienda4.plot(kind = 'barh', figsize = (12,5), color = 'blue')
```

```python
calificacion_tienda = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

resumen2 = pd.DataFrame({
                        'N° de Tienda' : [1, 2, 3, 4],
                        'Calificación por tienda' : [calificacion_tienda, calificacion_tienda2, calificacion_tienda3, calificacion_tienda4]
})

resumen2_ordenado = resumen2.sort_values('Calificación por tienda', ascending=False)
resumen2_ordenado
```

```python
resumen2_ordenado.plot(kind = 'bar', x = 'N° de Tienda', y = 'Calificación por tienda', figsize = (12,5), color = 'purple')
```

```python
tienda_promaventa['N° de Tienda'] = 1
tienda2_promaventa['N° de Tienda'] = 2
tienda3_promaventa['N° de Tienda'] = 3
tienda4_promaventa['N° de Tienda'] = 4

resumen3 = pd.concat([tienda_promaventa, tienda2_promaventa, tienda3_promaventa, tienda4_promaventa], ignore_index=True)
resumen3_ordenado = resumen3.sort_values('count', ascending=False)
resumen3_ordenado
```

```python
resumen3_ordenado.plot(kind = 'barh', x='Producto', y='count', figsize = (12,5), color = 'skyblue').invert_yaxis()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

palette = {1: '#363636', 2: '#E7133F', 3: '#FF83A5', 4: '#CCCCCC'}

plt.figure(figsize=(12,5))

sns.barplot(
    data=resumen3_ordenado,
    y='Producto',
    x='count',
    hue='N° de Tienda',
    orient='h',
    palette=palette
)
```
